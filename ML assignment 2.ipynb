```python
import pandas as pd
import seaborn as sns  
import matplotlib.pyplot as plt
df = pd.read_csv('diabetes_binary_classification_data.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
diabetes_counts = df['Diabetes_binary'].value_counts()
diabetes_counts
```

```python
df.corr()['Diabetes_binary'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Diabetes_binary', data=df)
plt.title('Number of Individuals with and without Diabetes')
plt.xticks(ticks=[0, 1], labels=['No Diabetes', 'Prediabetes/Diabetes'])
for i, count in enumerate(diabetes_counts):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=12)
plt.show()
```
